```python
movie = {
    "title": "Her",
    "genre": "Sci-Fi",
    "rating": 9.1
}

print(movie["title"])      # 키로 값 접근
print(movie["rating"])     # 평점 출력
```

```python
user = {
    "name": "홍길동",
    "age": 30,
    "skills": ["Python", "AI", "Docker"],   # 리스트 포함
    "profile": {"job": "Engineer", "level": "Senior"}   # 딕셔너리 포함
}
print(user["skills"])  
print(user["skills"][1])         # 리스트 내부 접근
print(user["profile"]["job"])    # 중첩 딕셔너리 접근
```

```python
movie = {"title": "Her", "rating": 9.1}

movie["genre"] = "Sci-Fi"      # 새 항목 추가
movie["rating"] = 9.3          # 값 수정
del movie["title"]             # 항목 삭제

print(movie)

movie["Tag"] = "sky"
print(movie)
```

```python
for key, value in movie.items():
    print(key, ":", value)
```

```python
import json

user = {"name": "Alice", "age": 25}
json_str = json.dumps(user, ensure_ascii=False) # 딕셔너리를 JSON 문자열로 변환한다.
print(json_str)
```
